```python
inBarcode='TCGA-DU-5870-10A-01D-1705-08'
barcodeToRDict={}
for inBarcode in tqdm(preprocessed_vcfs.index.unique()):
    #if inBarcode in preprocessed_vcfs.index:
    try :
        preprocessed_vcf_one=preprocessed_vcfs.loc[inBarcode]

        gdc_meta_df_sub_barcodeDf=gdc_meta_df_sub.set_index('BARCODE')

        raw_bam_uuid=gdc_meta_df_sub_barcodeDf.loc[inBarcode]['file_id']

        my_vcf_one=my_vcfs.loc[('TCGA',raw_bam_uuid)]

        #preprocessed_vcf_one.head()

        preprocessed_vcf_one[('one_vcf_df','CHROM')]=preprocessed_vcf_one[('one_vcf_df','CHROM')].str.replace('chr','')

        #RD, reference, AD: alternative
        preprocessed_vcf_one_index=preprocessed_vcf_one.set_index([('one_vcf_df','CHROM'),('one_vcf_df','POS'),('one_vcf_df','REF')])
        preprocessed_vcf_rdS=preprocessed_vcf_one_index[('vcf_info','RD')]#.values


        preprocessed_vcf_rdS.index.names=['CHROM','POS','base']

        preprocessed_vcf_one_index=preprocessed_vcf_one.set_index([('one_vcf_df','CHROM'),('one_vcf_df','POS'),('one_vcf_df','ALT')])

        preprocessed_vcf_adS=preprocessed_vcf_one_index[('vcf_info','AD')]
        preprocessed_vcf_adS.index.names=['CHROM','POS','base']

        preprocessed_vcf_S=pd.concat([preprocessed_vcf_rdS,preprocessed_vcf_adS])

        preprocessed_vcf_S=preprocessed_vcf_S.groupby(preprocessed_vcf_S.index.names).first()

        my_vcf_one_m=my_vcf_one.index.get_level_values('Pos').isin(
        preprocessed_vcf_S.index.get_level_values('POS'))
        preprocessed_vcf_S=preprocessed_vcf_S.astype(float)
        #my_vcf_one_m.mean()
        my_vcf_one=my_vcf_one[my_vcf_one_m].groupby(my_vcf_one.index.names).first()
        mergedDf=pd.concat([preprocessed_vcf_S[preprocessed_vcf_S>10],
                            my_vcf_one['ReadDepth']],axis=1).dropna()
        mergedDf=mergedDf#[mergedDf.ReadDepth>0]
        mergedDf.columns=['preprocessed','my']
        r,p=stats.pearsonr(mergedDf['preprocessed'],mergedDf['my'])
        #print (r)
        barcodeToRDict[inBarcode]={'R':r,'n':mergedDf.shape[0]}
    except KeyError:
        print (inBarcode)
```
